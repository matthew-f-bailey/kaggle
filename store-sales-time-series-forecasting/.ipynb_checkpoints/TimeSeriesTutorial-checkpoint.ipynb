```python
sns.scatterplot(data=oil, x="time", y = "dcoilwtico")
plt.xlabel("Days from Jan 2013")
plt.ylabel("Oil Price")
```

We are missing some dates here. Lets fill those in and impute the values

```python
# Add missing dates and backfill them
oil_dt = oil

# Covert to datetime and set index
oil_dt["date"] = pd.to_datetime(oil["date"], format="%Y-%m-%d")
oil_dt = oil_dt.sort_values(by=["date"], ascending=True)
oil_dt.set_index("date", inplace=True)
oil_filled = oil_dt.reindex(
    pd.date_range('01-01-2013','08-31-2017'), method="bfill"
)

# Fill in first missing value with backfill
bfil = oil_filled.iloc[1].dcoilwtico
oil_filled.iloc[0] = oil_filled.iloc[1]

# Add time index and date column
oil_filled["time"] = np.arange(len(oil_filled))
oil_filled["date"] = oil_filled.index

# Plot
sns.scatterplot(data=oil_filled, x="time", y = "dcoilwtico")
plt.xlabel("Days from Jan 2013")
plt.ylabel("Oil Price")
```

### Apply function in order to preprocess test as well

```python
def preprocess(df: pd.DataFrame, group_stores: bool = False):
    
    df["date"] = pd.to_datetime(df["date"], format="%Y-%m-%d")
    
    # Add store info for each listing
    df_feats = df.merge(stores, on="store_nbr")
    df_feats.rename(columns={"type":"store_type"}, inplace=True)
    
    # Add oil price on that day
    df_feats = df_feats.merge(oil_filled, on="date")
    df_feats["oil_price"] = df_feats["dcoilwtico"]
    df_feats.drop("dcoilwtico", axis=1, inplace=True)
    df_feats.drop(["time"], axis=1, inplace=True)
    
    # Add Holiday data for that day
    df_feats = df_feats.merge(merged_holidays, on="date", how="left")
    df_feats.rename(columns={"type":"holiday_type"}, inplace=True)
    
    # Check for local holiday
    df_feats['is_local_holiday'] = np.where(df_feats['city']==df_feats['locale_name'], 1, 0)
    df_feats['is_national_holiday'] = df_feats.apply(lambda x: 1 if "National" in str(x['locale']) and not x['transferred'] else 0, axis=1)
    
    df_feats.drop(["holiday_type","locale","locale_name","description","transferred", "state"],axis=1, inplace=True)
    
    # Add day of week
    df_feats['day_of_week'] = df_feats['date'].dt.day_name()
    df_feats['year'] = df_feats['date'].dt.year
    df_feats['day_of_year'] = df_feats['date'].dt.dayofyear
    
    df_feats = pd.get_dummies(df_feats, columns=["store_nbr", "family", "city", "store_type", "cluster", "day_of_week"], drop_first=True)

#     agg_methods = {
#         "family": "first",
#         "sales": "sum",
#         "onpromotion": "mean",
#         "city": "first",
#         "store_type": "first",
#         "cluster":"first",
#         "oil_price": "first",
#         "is_local_holiday":"first",
#         "is_national_holiday": "first",
#         "day_of_week": "first"
#     }
#     grouped = df_feats.groupby(["date","store_nbr"]).agg(agg_methods)
    
#     grouped = pd.get_dummies(grouped, columns=["family", "day_of_week", "city", "store_type", "cluster"], drop_first=True)
    
#     grouped = grouped.reset_index(level=['date','store_nbr'])
     
#     grouped = pd.get_dummies(grouped, columns=["store_nbr"], drop_first=True)
    
#     grouped["time"] = np.arange(len(grouped))
        
    return df_feats
```

```python
train_grouped = preprocess(train, True)
```

```python
X = train_grouped.drop(["id", "date", "sales"], axis=1)
y = train_grouped["sales"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
preds = lin_reg.predict(X_test)
print("Coefficients: \n", lin_reg.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_test, preds))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test, preds))
```
